MINI PROJECT : Student Data Explorer

```python

import pandas as pd
```

```python
df = pd.read_csv("student_performance.csv")
```

```python
df['total_score'  ] = df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score']
print("=" * 50)
print("      STUDENT PERFORMANCE EXPLORER")
print("=" * 50)

# 2. Dataset Overview
print("\nDATASET OVERVIEW")
print("-" * 30)

print(f"Total number of students : {len(df)}")
print(f"Total number of columns  : {len(df.columns)}")
print(f"Unique departments       : {df['department'].nunique()}")

# 3. Department Wise Analysis
print("\nDEPARTMENT WISE ANALYSIS")
print("-" * 30)

dept_count = df["department"].value_counts()

for dept, count in dept_count.items():
    print(f"{dept} : {count} students")

# 4. Score Analysis
print("\nSCORE ANALYSIS")
print("-" * 30)

print(f"Highest Math Score        : {df['math_score'].max()}")
print(f"Lowest Math Score         : {df['math_score'].min()}")
print(f"Average Math Score        : {df['math_score'].mean():.2f}")
print(f"Average Programming Score : {df['programming_score'].mean():.2f}")
print(f"Average Attendance        : {df['attendance_percentage'].mean():.2f}")

# 5. Gender Wise Analysis
print("\nGENDER WISE ANALYSIS")
print("-" * 30)

gender_avg = df.groupby("gender")["total_score"].mean()

for gender, avg in gender_avg.items():
    print(f"{gender} Average Total Score : {avg:.2f}")

# 6. Top Performers
print("\nTOP 5 PERFORMERS")
print("-" * 30)

top_students = df.sort_values(by="total_score", ascending=False).head(5)

for index, row in top_students.iterrows():
    print(f"Name       : {row['name']}")
    print(f"Department : {row['department']}")
    print(f"Total Score: {row['total_score']}")
    print("-" * 20)

# 7. Attendance Analysis
print("\nATTENDANCE ANALYSIS")
print("-" * 30)

low_attendance = df[df["attendance_percentage"] < 75]

if len(low_attendance) > 0:
    for index, row in low_attendance.iterrows():
        print(f"Name       : {row['name']}")
        print(f"Department : {row['department']}")
        print(f"Attendance : {row['attendance_percentage']}")
        print("-" * 20)
else:
    print("All students have attendance above 75%")
```
